```python
import librosa
import matplotlib.pyplot as plt
import librosa.display
import os
import numpy as np
from tensorflow.keras import layers
from IPython.display import Audio
import wave
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data_processing/sep28k-mfcc.csv")
df = df.reset_index()
df.head()
```

```python
df = df.drop(columns=['index'])
df.head()
```

```python
df = df[df['NaturalPause'] == 0]
df = df[df['Interjection'] == 0]
df = df[df['Prolongation'] == 0]
df = df[df['WordRep'] == 0]
df = df[df['Block'] == 0]
df.head()
```

```python
df = df.reset_index()
df = df.drop(columns=['index'])
```

```python
df
```

```python
df.loc[df['SoundRep'] >= 1.0, 'SoundRep'] = 1.0
df['SoundRep'].value_counts()
df
```

```python
df = df.reset_index()
```

```python
df
```

```python
df.drop(columns=['index'])
X = df.iloc[: , -39:]
X
```

```python
y = df['SoundRep']
y.to_csv("test.csv",index=False)
X.to_csv("ysy.csv", index=False)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.2, random_state=12)
```

```python
print(len(X_train), len(X_test), len(y_train), len(y_test))
```

```python
from collections import Counter
print(Counter(y_train))
print(Counter(y_test))
```

```python
def build_model(input_shape, num_classes):
    inputs = tf.keras.Input(shape=input_shape)
    x = layers.Dense(39, activation="relu")(inputs)
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(13, activation="relu")(x)

    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = tf.keras.Model(inputs=inputs, outputs=outputs)

    return model
```

```python
input_shape = (39,)

num_classes = 2  # for binary classification (yes/no)
model = build_model(input_shape, num_classes)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=16, epochs=20)
```

```python
loss, accuracy = model.evaluate(X_train, y_train)
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')
```

```python
predictions = model.predict(X_test, batch_size=123)
predicted_class = tf.argmax(predictions, axis=1)
print(predicted_class)

print(Counter(np.array(predicted_class)))
print(Counter(np.array(y_test)))
```
